```python
# ===== STEP 1: LOAD FILE =====
with open("posts.txt", "r", encoding="utf-8") as f:
    data = f.read()

# ===== STEP 2: SPLIT POSTS =====
posts = data.split("---")

# ===== STEP 3: CLEAN POSTS =====
clean_posts = []
for p in posts:
    text = p.strip()
    if len(text) > 50:   # remove noise/empty
        clean_posts.append(text)

print("Total clean posts:", len(clean_posts))
print("\nSample:\n", clean_posts[0][:300])


# ===== STEP 4: STRUCTURE DATA =====
import pandas as pd

df = pd.DataFrame({
    "post": clean_posts
})

print("\nDataFrame preview:")
print(df.head())


# ===== STEP 5: SAVE TO CSV =====
df.to_csv("posts.csv", index=False)

print("\nSaved as posts.csv")
```
